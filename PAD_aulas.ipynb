# **PAD 2.1 Introdução aos arrays**



numpy arrays são:
* Estruturas de dados para manipulação e álgebras matricial.
* Possibilita trabalhar com estruturas de dados n-dimensionais.

1D array:
* possui apenas o axis 0 (horizontal)
* direção ao longo das linhas.
* shape: (4,)
* como é 1D o número 4 representa a quantidade de elementos do array.

2D array:
* possui axis 0 (horizontal) e axis 1 (vertical)
* direção ao longo das linhas e colunas
* shape: (2, 3)
* como é 2D o número 2 é a quantidade de linhas e 3 a quantidade de colunas.

3D array:
* possui axis 0 (horizontal), axis 1(vertical) e axis 2 (profundidade)
* direção ao longo dos eixos x, y e z. (ex.: imagens).
* shape: (4, 3, 2)
* como é 3D o número 4 é a quantidade de linhas, o 3 é a quantidade de colunas e o 2 é a quantidade de camadas(profundidade).

# **PAD 2.3 Indexação de arrays**

Os índices ao longo de uma dimensão variam de `0 a n-1`

`n` é o número de elementos da dimensão

```
index   0    1  2   3   -1
A    =  88  19  46  74  94
```
Acessando direto no índice: `A[2]=46`
Último elemento: `A[-1]=94` ou `A[4]=94`

* usando números negativos é contado de trás para frente então o último elemento é -1, o penúltimo é -2 e assim por diante.

image.png

Acessar elementos em um array 2D (matriz B):

`B[i, j]`
* `i` acessa linha do array (axis 0)
* `j` acessa coluna do array (axis 1)


```
index   0    1  2   3   -1
A    =  88  19  46  74  94
```
**Slicing**

Acessando mais de um elemento em um array: 

`A[1:3]` 
* `1` é o primeiro íncide (inclusivo)
* `3` é o último íncide (exclusivo)

Vai retornar: `19, 46`


image.png

**Slicing**

Acessar elementos em um array 2D (matriz B):

`B[i:k, j:l]`
* os índices `k`e `l` **não** entram no slicing
* o slicing incluirá até `k-1` e `l-1`, respectivamente


# **PAD 2.4 Indexação de arrays - Prática**

**Indexação de arrays**

````
# os índices no Python vão de 0 a n-1,
# onde n é o tamanho da dimensão
x = np.linspace(start=10, stop=100, num=10)
print('x:', x)
print('shape:', x.shape)
````

````
# extração de elementos
print('x:', x)
print('primeiro elemento:', x[0])
print('segundo elemento:', x[1])
print('último elemento:', x[9])
print('último elemento:', x[-1])
````

````
# slicing: extração de subarrays:
print('x:', x)
print('dois primeiros elementos:', x[0:2])  # 2 é exclusivo
print('dois primeiros elementos:', x[:2]) # iniciando com : é a partir do primeiro elemento
print('dois últimos elementos:', x[-2:]) # : seguido de nada é até o final
````

````
# slicing em arrays 2D (matrizes)
# redimensionando o array para 2D
x = x.reshape(2,5) # reshape de x para 2 linhas e 5 colunas
print('x:\n', x)
````

````
# extração de elementos
print('primeira linha, segunda coluna:', x[0, 1])
print('segunda linha, penúltima coluna:', x[1, -2])
print('última linha, última coluna:', x[1, 4])
print('última linha, última coluna:', x[-1, -1])
````

````
print(x)
````

````
# slicing: extração de subarrays
print('x:\n', x)
print('primeira linha inteira: ', x[0, :]) # o : sem nada antes e no final significa que via pegar TODOS elementos
print('primeira linha, segunda a quarta coluna: ', x[0, 1:4])
print('última coluna inteira:\n', x[:, [-1]])  # o [] preserva a dimensão original
````

````
# atenção com compartilhamento de memória em subarrays!!
x = np.array([1, 2, 3])
print('x antes:', x)
y = x[:2]
y[0] = -100  # alteração do valor em y altera o valor de x
print('x depois:', x)
````

````
# atenção com compartilhamento de memória em subarrays!!
x = np.array([1, 2, 3])
print('x antes:', x)
y = x[:2].copy()  # faz uma cópia e não compartilha a memória
y[0] = -100  # alteração do valor em y NÃO altera o valor de x
print('x depois:', x)
````

# **PAD 2.7 Operações aritméticas multiplicação: element wise e matricial - Prática**

````
# criação de dois arrays x e y
x = np.ones((2, 2))
y = np.eye(2)
print('x: \n', x)
print('y: \n', y)
````

````
# multiplicação elemento a elemento
print('x: \n', x)
print('y: \n', y)
````

````
# multiplicação
print('multiplicação de dois arrays:\n', x * y)
print('multiplicação com float/int:\n', x * 2)  # broadcasting
````

````
# multiplicação matricial
print('multiplicação matricial (np.dot):\n', np.dot(x, y))
print('multiplicação matricial (@):\n', x @ y)  # método nativo do python
print('multiplicação matricial (.dot):\n', x.dot(y))
````

````
'''
Exemplo:
Solução de um sistema de equações:
    a + 2*b = 7
  3*a - 2*b = -11  
  solução análitica: (a, b) = (-1, 4)
Matricialmente, este problema tem a seguinda forma:
  Ax = c, onde:
  - x = [a, b]
  - A = [[1, 2], [3, -2]]
  - c = [7, -11]
  solução numérica: x = inv(A) @ c,
'''

# definição do problema
A = np.array([[1, 2], [3, -2]])
c = np.array([[7], [-11]])
print('A: \n', A)
print('c: \n', c)
````

**Sistema de equações:**

image.png

````
#diversos métodos para algebra linear a partir do método abaixo
#np.linalg.
````

````
# solução
x = np.dot(np.linalg.inv(A), c)
#x = np.linalg.inv(A) @ c
print('(a, b):', x.ravel())
````

# **2.8 Comparações e Indexação booleana**

````
#importando o numpy com o apelido np
import numpy as np
````

**Comparações**

Comparação **menor/menor** ou **igual**:

````
# comparações booleanas
A = np.array([1, 2, 3])
B = np.array([2, 0, 2])
s = 3

# menor
print('\nComparação menor:')
print(A < B)
print(A < s)

# menor ou igual
print('\nComparação menor ou igual:')
print(A <= B)
print(A <= s)
````

Comparação **maior/maior** ou **igual**:

````
# comparações booleanas
A = np.array([1, 2, 3])
B = np.array([2, 0, 2])
s = 3

# maior
print('\nComparação maior:')
print(A > B)
print(A > s)

# maior ou igual
print('\nComparação maior:')
print(A >= B)
print(A >= s)
````

**Igualdade**:

````
# comparações booleanas
A = np.array([1, 2, 3])
B = np.array([2, 0, 2])
s = 3

# igual
print('\nComparação igualdade:')
print(A == B)
print(A == s)
````

**Indexação booleana**

Operação de filtro:

````
# indexação booleana: um novo subarray contendo uma
# cópia dos elementos em que a condição de verificação se aplica
cond = A <= 2
D = A[cond]
print('A:', A)
print('condição:', cond)
print('D:', D)
````

# **2.9 Comparações e Indexação booleana - Prática**

````
#importando o numpy com o apelido np
import numpy as np
````

**Comparações**

````
# criação dos arrays
x = np.array([[1, 2], [3, 4]])
y = np.array([1.5, 3.5])
print('x: \n', x)
print('y: \n', y)
````

````
# comparações ponto a ponto
print('Comparação de um array com um escalar (>): \n', x > 2)
print('Comparação de um array com um escalar (>=): \n', x >= 2)
print('Comparação de um array com um escalar (<): \n', x < 2)
print('Comparação de um array com um escalar (<=): \n', x <= 2)
print('Comparação entre arrays (==): \n', x == x)
print('Comparação entre arrays (>): \n', x > x)
print('Comparação entre arrays (>): \n', x > y) # broadcasting
````

**Indexação booleana**

````
# indexação booleana
x = np.array([[1, 3, 7],
             [4, 11, 21],
             [42, 8, 9]])
print('x:\n', x)
````

````
# indexação booleana: retornar o número de elementos
# maiores que k
k = 10
cond = x > k
print('condição: \n', cond)
print(f'elementos maiores que {k}:', x[cond])
print(f'número de elementos maiores que {k}:', len(x[cond]))
````

````
# indexação booleana: extração dos números pares
cond = x % 2 == 0 # números pares
print('condição: \n', cond)
print('números pares:', x[cond])
````

````
# indexação booleana: extração dos números ímpares
cond = x % 2 == 1 # números ímpares
print('condição: \n', cond)
print('números ímpares:', x[cond])
````
